```python
#data where the profit is negative, sorted by Profit
d2=round(d1[d1['Profit']<0].sort_values(by='Profit',ascending=True),2)
d2
```

```python
#when the profit is largest and smallest in negative
d2[['Ship Mode','Sales','Profit']][(d2['Profit']==-6599.98) | (d2['Profit']==-0.09)]
```

```python
#percentage when Ship Mode is Standard Class
f=len(d2[d2['Ship Mode']=='Standard Class'].value_counts().reset_index())
round((f/1871)*100,2)
```

```python
#percentage when Ship Mode is First Class
f=len(d2[d2['Ship Mode']=='First Class'].value_counts().reset_index())
round((f/1871)*100,2)
```

```python
#percentage when Ship Mode is Second Class
f=len(d2[d2['Ship Mode']=='Second Class'].value_counts().reset_index())
round((f/1871)*100,2)
```

```python
#percentage when Ship Mode is Same Day
f=len(d2[d2['Ship Mode']=='Same Day'].value_counts().reset_index())
round((f/1871)*100,2)
```

```python
l1=['Standard Class','Second Class','First Class','Same Day']
l2=[62.64,16.35,15.66,5.24]
```

```python
#Percentage of Ship Mode used
plt.figure(figsize=(14,6))
plt.pie(l2,labels=l1,colors=['Blue','Orange','Green','Yellow'],autopct='%0.1f%%')
plt.show()
```

```python
#taking top 20 values sorted by sales
b1=d2.nlargest(20,columns='Sales')
b1
```

```python
#counting total count of Standard Class in top 20 values
b1['Ship Mode'].value_counts()
```

```python
#graph showing ship modes and their count
plt.figure(figsize=(8,5))
plt.bar(b1['Ship Mode'].value_counts().keys(),b1['Ship Mode'].value_counts(),color='Green')
plt.show()
```

```python
#Bar Graph for top 20 values showing different Ship Modes, sorted by Sales
plt.figure(figsize=(6,5))
plt.xlabel('Ship Modes',labelpad=10)
plt.ylabel('Sales',labelpad=10)
plt.bar(b1['Ship Mode'],b1['Sales'])
plt.show()
```

```python
#analysing top 20 records where the Ship Mode is Standard Class
b1[['Profit','Ship Mode','Discount','Sales']][b1['Ship Mode']=='Standard Class'].reset_index()
```

```python
#Graph showing Negative Profit(loss) for different Ship Modes
plt.bar(b1['Ship Mode'],b1['Profit'],color='red')
plt.show()
```

```python
#plot between Dicount and Ship Mode for top 20 values
b2=b1['Ship Mode'][b1['Ship Mode']=='Standard Class']
b1.plot(kind='bar',x='Ship Mode',y='Discount',figsize=(14,6),color='brown')
plt.xlabel('Ship Mode',labelpad=20)
plt.ylabel('Discount',labelpad=20)
plt.show()
```
